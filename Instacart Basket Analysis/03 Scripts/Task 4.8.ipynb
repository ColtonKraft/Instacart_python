```python
ords_prods_merge.head(15)
```

```python
# Creating max order for entire dataframe
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Designating flags for customer loyalty
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking value counts for loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# Q5

```python
# Finding the average amount of orders by loyalty status
ords_prods_merge.groupby('loyalty_flag').agg({'order_number': ['mean']})
```

```python
# Finding average price of product for each loyalty status
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# Q5
It looks like new, regular, and loyal customers all buy products around the same price (within 0.03)

# Q6

```python
# Creating the spending flag
ords_prods_merge['user_prices_mean'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Assigning spending flags for each customer
ords_prods_merge.loc[ords_prods_merge['user_prices_mean'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['user_prices_mean'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head(20)
```

```python
# checking totals of high spenders vs. low spenders
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Q7

```python
# Creating median values for customers days since prior order column
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(5)
```

```python
# Creating frequency flags for each customer
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent Customer'                    
```

```python
# Checking frequency flags 
ords_prods_merge.head(5)
```

```python
# Frequency flags value counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
